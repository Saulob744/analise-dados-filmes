## üßπ Tratamento dos Dados

```python
import pandas as pd
notes = pd.read_csv(r"C:\Users\Administrador\Downloads\credits.csv\ratings.csv")
filme = pd.read_csv(r"C:\Users\Administrador\Downloads\credits.csv\movies.csv")
links = pd.read_csv(r"C:\Users\Administrador\Downloads\credits.csv\links.csv")
tags = pd.read_csv(r"C:\Users\Administrador\Downloads\credits.csv\tags.csv")
filmes = notes.merge(filme,on='movieId')
filmes = filmes.merge(filme,on='movieId')
filmes = filmes.merge(links,on='movieId')
filmes = filmes.merge(tags,on='movieId')
filmes = filmes[['movieId','rating','title_x','genres_x','imdbId']]
filmes
```

## üé¨ Top 10 Melhores Filmes Segundo as Avalia√ß√µes dos Usu√°rios

```python
melhores_filmes = filmes.groupby('title_x')[['title_x','rating']].sum().sort_values(by='rating',ascending=False)
melhores_filmes = melhores_filmes.rename(columns={'title_x': 'Nome','rating': 'Avaliacao'})
melhores_filmes[:10]
```

# üíÄ Piores 10 Filmes Segundo as Avalia√ß√µes dos Usu√°rios

```python
melhores_filmes[-10:].sort_values(by='Avaliacao')
```
